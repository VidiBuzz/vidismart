```
import collections
import collections.abc
import sys

# --- COMPATIBILITY PATCH START ---
# This fixes the common crash on Python 3.10+ where python-pptx fails
# because 'collections.Container' was moved to 'collections.abc.Container'
if not hasattr(collections, 'Container'):
    collections.Container = collections.abc.Container
if not hasattr(collections, 'Mapping'):
    collections.Mapping = collections.abc.Mapping
if not hasattr(collections, 'MutableMapping'):
    collections.MutableMapping = collections.abc.MutableMapping
# --- COMPATIBILITY PATCH END ---

try:
    from pptx import Presentation
    from pptx.util import Inches, Pt
    from pptx.dml.color import RGBColor
    from pptx.enum.text import PP_ALIGN
except ImportError:
    print("Error: 'python-pptx' library is missing.")
    print("Please run: pip install python-pptx")
    sys.exit(1)

# Create a presentation object
prs = Presentation()

# --- HELPER FUNCTIONS ---

def format_slide(slide, title_text):
    """Sets the background to white and formats the title."""
    background = slide.background
    fill = background.fill
    fill.solid()
    fill.fore_color.rgb = RGBColor(255, 255, 255)

    if slide.shapes.title:
        title = slide.shapes.title
        title.text = title_text

        # Style Title (Mimicking Kumbh Sans with Arial)
        if title.text_frame.paragraphs:
            p = title.text_frame.paragraphs[0]
            p.font.name = 'Arial'
            p.font.size = Pt(40)
            p.font.bold = True
            p.font.color.rgb = RGBColor(0, 0, 0) # Black
            p.alignment = PP_ALIGN.LEFT

            # Position title
            title.top = Inches(0.5)
            title.left = Inches(0.5)
            title.width = Inches(9)
            title.height = Inches(1.5)

def add_bullets(slide, items):
    """Adds a text box with bullet points to the slide."""
    left = Inches(0.5)
    top = Inches(1.8)
    width = Inches(9)
    height = Inches(5)

    txBox = slide.shapes.add_textbox(left, top, width, height)
    tf = txBox.text_frame
    tf.word_wrap = True

    for item in items:
        p = tf.add_paragraph()
        p.text = item
        p.font.name = 'Arial'
        p.font.size = Pt(24)
        p.font.color.rgb = RGBColor(51, 51, 51) # Dark Grey
        p.space_after = Pt(14)
        p.level = 0

# --- SLIDE GENERATION ---

# Slide 1: Title
slide_layout = prs.slide_layouts[0] # Title Slide
slide = prs.slides.add_slide(slide_layout)
title = slide.shapes.title
subtitle = slide.placeholders[1]

title.text = "VidiSmart AI"
subtitle.text = "The Architecture of Veracity & The Ai-SmartStack\n\nJames May\n561.510.0021 | https://VidiSmart.com"

# Style Title Slide
if title.text_frame.paragraphs:
    title.text_frame.paragraphs[0].font.bold = True
    title.text_frame.paragraphs[0].font.color.rgb = RGBColor(0, 86, 179) # Accent Blue

if subtitle.text_frame.paragraphs:
    subtitle.text_frame.paragraphs[0].font.color.rgb = RGBColor(80, 80, 80)

# Slide 2: Trust Gap
slide_layout = prs.slide_layouts[5] # Blank
slide = prs.slides.add_slide(slide_layout)
format_slide(slide, "The Crisis of Confidence")
add_bullets(slide, [
    "The Problem: Enterprises face a 'trust gap' throttling widespread AI adoption.",
    "Root Cause: Current LLMs act as 'stochastic parrots'—predicting probabilities rather than verifying truth.",
    "The Consequence: When hallucinations occur, the 'Trust Loop' breaks.",
    "The Requirement: A system that produces a 'Rational Response'—semantically relevant and factually cited."
])

# Slide 3: Solution
slide = prs.slides.add_slide(slide_layout)
format_slide(slide, "VidiSmart: The Architecture of Veracity")
add_bullets(slide, [
    "Unifying Five Modalities to Close the Loop:",
    "1. Vector: Encodes semantic meaning and intent (The 'Why').",
    "2. Lexical: Ensures exact keyword matching (The 'What').",
    "3. Graph: Validates logical consistency (The 'How').",
    "4. Geospatial: Anchors data to physical reality (The 'Where').",
    "5. Visual: Provides perceptual certainty (The 'Show Me')."
])

# Slide 4: Core Engine
slide = prs.slides.add_slide(slide_layout)
format_slide(slide, "The Ai-SmartStack")
add_bullets(slide, [
    "Powered by Vidi.ai",
    "Omni-Hybrid Capability: Lexical, vector, structured, and tensor computation simultaneously.",
    "Compute Over Data: Algorithms run directly on the data node, eliminating latency.",
    "Proven Scale: Built on the stack that powered Yahoo/Flickr at billion-scale.",
    "Why it wins: Transcends simple 'vector databases' to become a true AI Search Platform."
])

# Slide 5: Visual AI
slide = prs.slides.add_slide(slide_layout)
format_slide(slide, "Visual AI & The \"Asset Brain\"")
add_bullets(slide, [
    "Visual Grounding: Using image/video input eliminates text-based ambiguity.",
    "The 'Show Me' Response: Generating visuals reduces time-to-decision.",
    "Production Pipeline: Acts as the 'Asset Brain' for tools like ComfyUI.",
    "Continuous Learning: Generated assets create a stateful, learning loop."
])

# Slide 6: Orchestration
slide = prs.slides.add_slide(slide_layout)
format_slide(slide, "Agentic Orchestration")
add_bullets(slide, [
    "True Federation: The federator is a dynamic Agent Workflow, not a database.",
    "Dynamic Planning: Agents decompose complex queries into sub-tasks.",
    "The Toolbox:",
    "- OpenSearch: Primary hybrid store for unstructured text.",
    "- Trino: Handles live, federated SQL queries.",
    "- Visual Vector: Manages high-performance tensor/vector operations."
])

# Slide 7: Feedback
slide = prs.slides.add_slide(slide_layout)
format_slide(slide, "Adaptive Feedback Loops")
add_bullets(slide, [
    "The Loop: Pairing Visual Vector (serving) with ClickHouse (analytics).",
    "Self-Optimizing: Aggregates user clicks in real-time.",
    "Adaptive Ranking: Search results adjust to trends instantly.",
    "System-Level RLHF: Continuous reinforcement learning from human feedback."
])

# Slide 8: Security
slide = prs.slides.add_slide(slide_layout)
format_slide(slide, "Security & Sovereignty")
add_bullets(slide, [
    "Data Sovereignty: Supports air-gapped, on-premise deployments.",
    "Open Source Models: Replaces 'Black Box' APIs with local models (CLIP, ColPali).",
    "Security First: Full support for mTLS encryption and RBAC.",
    "Cost Control: Lower TCO by eliminating per-call API fees."
])

# Slide 9: Conclusion
slide = prs.slides.add_slide(slide_layout)
format_slide(slide, "The Future of Trusted AI")
add_bullets(slide, [
    "The VidiSmart Advantage: Moving beyond the 'database + plugin' approach.",
    "Closing the Loop: Verifying the What, Why, How, Where, and Appearance.",
    "Strategic Vision: Designed for the next decade of autonomous systems."
])

# Slide 10: Contact
slide = prs.slides.add_slide(slide_layout)
format_slide(slide, "Connect with VidiSmart")
add_bullets(slide, [
    "James May",
    "561.510.0021",
    "https://VidiSmart.com"
])

# Save the presentation
output_path = 'VidiSmart_Presentation.pptx'
prs.save(output_path)
print(f"Successfully generated: {output_path}")
```
